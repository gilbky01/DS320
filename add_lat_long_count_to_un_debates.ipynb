```python
import requests
import csv
```

```python
res = requests.get("https://opendata.socrata.com/api/views/mnkm-8ram/rows.json?accessType=DOWNLOAD")
```

```python
res.status_code
```

```python
j = res.json()
```

```python
print(j.keys())
print(j['data'][10])
```

```python
latlong = {}

for i in j['data']:
    latlong[i[10]] = {}
    latlong[i[10]]['long'] = i[13]
    latlong[i[10]]['lat'] = i[12]
    
# for latitude, negative numbers mean south, and for longitude negative numbers mean west
        
print(latlong)
```

```python
with open('/home/faculty/millbr02/pub/un-debates-augmented-1.csv', 'r', encoding='latin-1') as f:
    csv_iter = csv.reader(f)
    db = []
    for row in csv_iter:
        db.append(row)
        
print(db[0])

db[0].append('latitude')
db[0].append('longitude')
print(db[0])
```

```python
for rec in db:
    rec[3] = rec[3].replace('\n', '').replace(',', '')
    for i in latlong:
        if rec[2] == i:
            rec.append(latlong[i]['lat'])
            rec.append(latlong[i]['long'])

print(db[0:3])
```

```python
with open('un_debates_with_lat_long.csv', 'w', encoding='utf-8') as out:
    for rec in db:
        if len(rec) == 9:
            out.write(rec[0] + ', ' + rec[1] + ', ' + rec[2] + ', ' + rec[3] + ', ' + rec[4] + ', ' + rec[5] + ', '
                     + rec[6] + ', ' + rec[7] + ', ' + rec[8] + '\n')
```

```python
mentions = {}
countries = []

for rec in db[1:]:
    if rec[4] not in countries:
        countries.append(rec[4])
    if rec[4] not in mentions:
        mentions[rec[4]] = 0
        
for rec in db[1:]:
    for country in countries:
        num = rec[3].count(country)
        mentions[country] += num
        
print(mentions)
```

```python
smaller_db = []
info = ['country_code', 'country_name', 'mentions', 'latitude', 'longitude']
smaller_db.append(info)
used_countries = []
    
for rec in db[1:]:
    num_mentions = mentions[rec[4]]
    if rec[4] not in used_countries and len(rec) == 9:
        smaller_rec = [rec[2], rec[4], num_mentions, rec[7], rec[8]]
        smaller_db.append(smaller_rec)
        used_countries.append(rec[4])
```

```python
print(smaller_db[0:10])
```

```python
with open('smaller_un_debates_lat_long_mentions.csv', 'w', encoding='utf-8') as out:
    for rec in smaller_db[0:1000]:
        out.write(rec[0] + ', ' + rec[1] + ', ' + str(rec[2]) + ', ' + rec[3] + ', ' + rec[4] + '\n')
```
