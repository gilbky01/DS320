```python
# don't run cells more than once otherwise the numbers will be wrong
# if you want to re-run a cell, restart this one then do the cell you want to run

import csv

with open('/home/faculty/millbr02/pub/un-debates-augmented-1.csv','r',encoding='latin-1') as f:
    csv_iter = csv.reader(f)
    db = []
    for row in csv_iter:
        db.append(row)
        
countries = []
years = []
mentions = {}
mentions_by_year = {}
mentions_by_country = {}

# add all countries and years to their lists and populate dictionaries
for rec in db[1:]:
    if rec[1] not in years:
        years.append(rec[1])
    if rec[4] not in countries:
        countries.append(rec[4])
        
    if rec[4] not in mentions_by_country:
        mentions_by_country[rec[4]] = {}
    if rec[4] not in mentions:
        mentions[rec[4]] = 0
    if rec[1] not in mentions_by_year:
        mentions_by_year[rec[1]] = {}
        
# finish populating mentions_by_year    
for year in years:
    for country in countries:
        if country not in mentions_by_year[year]:
            mentions_by_year[year][country] = 0

# finish populating mentions_by_country
for country1 in countries:
    for country2 in countries:
        if country2 not in mentions_by_country[country1]:
            mentions_by_country[country1][country2] = 0
        if country1 not in mentions_by_country[country2]:
            mentions_by_country[country2][country1] = 0


print("Done")
```

```python
# total mentions across all years

for rec in db[1:]:
    for country in countries:
        num = rec[3].count(country)
        mentions[country] += num
        
print("Total mentions:")
for country in mentions:
    print(country + ' was mentioned ' + str(mentions[country]) + ' times by others')
print()
```

```python
# mentions by year

for rec in db[1:]:
    for country in countries:
        num = rec[3].count(country)
        
        year = rec[1]
        mentions_by_year[year][country] += num
        
for year in mentions_by_year:
    print(year + ':')
    for country in mentions_by_year[year]:
        print(country + ' was mentioned ' + str(mentions_by_year[year][country]) + ' times by others in ' + year)
    print()
```

```python
# mentions by country across all years

for rec in db[1:]:
    for country in countries:
        num = rec[3].count(country)
        
        mentions_by_country[rec[4]][country] += num

for country1 in mentions_by_country:
    print("Mentions by " + country1 + ':')
    for country2 in mentions_by_country[country1]:
        print(country2 + ' was mentioned ' + str(mentions_by_country[country1][country2]) + ' times by ' + country1)
    print()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
import warnings
import math
warnings.filterwarnings("ignore")
```

```python
g = nx.Graph()
#g.add_nodes_from(countries)
```

```python
for country1 in mentions_by_country:
    for country2 in mentions_by_country[country1]:
        if mentions_by_country[country1][country2] > 50:
            g.add_node(country1)
            g.add_node(country2)
            g.add_edge(country1, country2, weight=mentions_by_country[country1][country2])
```

```python
d = nx.degree(g)
gfig, gax = plt.subplots(figsize=(15,15))
hego = nx.ego_graph(g, 'United States', radius=2)
pos = nx.spring_layout(hego, k=20*1/math.sqrt(len(mentions_by_country)), iterations=20)
nx.draw(hego, pos, ax=gax, node_size=300, with_labels=True, font_size=14, font_color='darkgreen')

#nx.draw_circular(g, with_labels=True, node_size=[v * 0.25 for v in d.values()], font_size=10, font_color='blue')
plt.savefig("graph.png", dpi=75)
```
